# Date and Time Variables -Feature Engineering

## Date Variables - Lending Club Loan Data -Kaggle

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
loan=pd.read_csv('D:/ShakthiSai -Keerthana Learnings/AI and Machine Learning/Feature Engineering/loan.csv',usecols=['issue_d','last_pymnt_d'])
loan.head()
```

```python
loan.info()
```

```python
## strings to datetime format

loan['issue_date']=pd.to_datetime(loan['issue_d'])
loan['last_pymnt_date']=pd.to_datetime(loan['last_pymnt_d'])
```

```python
loan.head()
```

```python
## extract year

loan['issue_d_year']=loan['issue_date'].dt.year
loan.head()
```

```python
loan['issue_d_year'].unique()
```

```python
## extract week of the year - will return values from 1 to 52

loan['issue_d_week']=loan['issue_date'].dt.week
loan.head()
```

```python
loan['issue_d_week'].unique()
```

```python
## extract month of the year - will return values from 1 to 12

loan['issue_d_month']=loan['issue_date'].dt.month
loan.head()
```

```python
loan['issue_d_month'].unique()
```

```python
## extract quarter of the year - will return values from 1 to 4

loan['issue_d_quarter']=loan['issue_date'].dt.quarter
loan.head()
```

```python
loan['issue_d_quarter'].unique()
```

```python
## extract as per semesters
## sem 1: q1,q2
## sem 2: q3,q4

loan['issue_d_sem']=np.where(loan['issue_d_quarter'].isin([1,2]),1,2)
loan.head()
```

```python
loan['issue_d_sem'].unique()
```

```python
## extract days on month 1-31

loan['issue_d_day']=loan['issue_date'].dt.day
loan.head()
```

```python
loan['issue_d_day'].unique()
```

```python
## extract day of week- mon to sun &  0 to 6

loan['issue_d_dayofweek']=loan['issue_date'].dt.dayofweek
loan['issue_d_dayofweekname']=loan['issue_date'].dt.day_name()
loan.head()
```

```python
loan['issue_d_dayofweek'].unique()
```

```python
loan['issue_d_dayofweekname'].unique()
```

```python
## weekend?

loan['issue_d_weekend']=np.where(loan['issue_d_dayofweekname'].isin(['Saturday','Sunday']),1,0)
loan.head()
```

```python
## extract difference between dates
(loan['last_pymnt_date']-loan['issue_date']).dt.days
```

```python
## months passed between 2 dates
## The arguments for timedelta64 are a number, to represent the number of units, and a date/time unit, such as (D)ay, (M)onth, (Y)ear, (h)ours, (m)inutes, or (s)econds. The timedelta64 data type also accepts the string “NAT” in place of the number for a “Not A Time” value.


loan['months_passed']=(loan['last_pymnt_date']-loan['issue_date'])/np.timedelta64(1,'M')
loan['months_passed']=np.round(loan['months_passed'],0)
loan.head()
```
