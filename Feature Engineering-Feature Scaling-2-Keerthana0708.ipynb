# Feature Scaling

## Techniques : 
   ### 1)Robust Scaling
   ### 2)Scaling to vector unit length (Scaling to unit norm)

## Robust Scaling- Boston Dataset

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
bos=load_boston()
```

```python
bos
```

```python
bosds=pd.DataFrame(bos.data,columns=bos.feature_names)
bosds['MEDV']=bos.target
bosds.head()
```

```python
## TT split
```

```python
X=bosds.iloc[:,:-1]
y=bosds['MEDV']
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
## Robust Scaling
```

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
rs.fit(X_train)  ## fit the scalar to train ,it will learn the parameters
```

```python
## scaler stores the median values of the features as learned from train set
rs.center_
```

```python
## scaler stores the median values of the features as learned from train set
rs.scale_
```

```python
## transform the train and test

X_train_scaled=rs.transform(X_train)
X_test_scaled=rs.transform(X_test)
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
## median of original ds
## np.round to reduce the no of decimals to 1
np.round(X_train.median(),1)
```

```python
np.round(X_train_scaled.median(),1)
```

## Scaling to vector unit length (Scaling to unit norm)

```python
from sklearn.preprocessing import Normalizer
```

```python
## Scaling to L1
```

```python
sc=Normalizer(norm='l1')
sc.fit(X_train)
X_train_scale=sc.transform(X_train)
X_test_scale=sc.transform(X_test)
```

```python
## calculate the manhattan distance observation per observation

np.round(np.linalg.norm(X_train_scale,ord=1,axis=1),1)
```

```python
## scaled data

np.round(np.linalg.norm(X_train_scale,ord=1,axis=1),1)
```

```python
## now each feature vector has l1 norm of 1
```

```python
X_train_scale=pd.DataFrame(X_train_scale,columns=X_train.columns)
X_test_scale=pd.DataFrame(X_test_scale,columns=X_test.columns)
```

```python
X_train.describe()
```

```python
X_train_scale.describe()
```

```python
## Scaling to l2
```

```python
sca=Normalizer(norm='l2')
sca.fit(X_train)
X_train_scale=sca.transform(X_train)
X_test_scale=sca.transform(X_test)
```

```python
## calculate the euclidean distance observation per observation

np.round(np.linalg.norm(X_train,ord=2,axis=1),1)
```

```python
np.round(np.linalg.norm(X_train_scale,ord=2,axis=1),1)
```
